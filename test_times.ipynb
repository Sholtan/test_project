```python
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
```

#Data generation

```python
np.random.seed(42)
days = np.arange(30).reshape(-1, 1)
values = 10 + 0.5 * days.flatten() + np.random.normal(scale=1, size=30)
```

#Training model

```python
model = LinearRegression()
model.fit(days, values)
```

#Prediction

```python
future_days = np.arange(30, 50).reshape(-1, 1)
predictions = model.predict(future_days)

# Plot
```

```python
plt.figure(figsize=(8, 4))
plt.plot(days, values)
plt.plot(future_days, predictions)
plt.title("Synthetic Time Series Forecast")
plt.xlabel("Day")
plt.ylabel("Value")
plt.grid(True)
plt.show()

predictions
```
